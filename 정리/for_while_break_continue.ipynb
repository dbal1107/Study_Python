```python
for i in range(2010, 2023):
    for j in range(1, 13):
        print(f"{i}-{str(j).zfill(2)}")
```

```python
for i in range(2010, 2023, 2):
    for j in range(2, 14, 2):
        print(f"{i}-{str(j).zfill(2)}")
```

```python
for i in range(2010, 2023):
    for j in range(1, 13):
        if i%2==0 and j%2==0:
            print(f"{i}-{str(j).zfill(2)}")
```

```python
for i in range(5):
    for j in range(5):
        print(f'{"."*5-i}{"+"*j-i}')
```
