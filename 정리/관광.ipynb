```python
kto_total = pd.DataFrame()
continents_list = ['아시아주','미주','구주','대양주','아프리카주','기타대륙','교포소계']
continents = ['아시아'] * 25 + ['아메리카'] * 5 + ['유럽'] * 23 + ['오세아니아'] * 3 + ['아프리카'] * 2 + ['기타대륙'] + ['교포']

for yy in range(2010, 2020):
    for mm in range(1, 13):
        try:
            yymm = '{}{}'.format(yy, str(mm).zfill(2))
            kto = pd.read_excel(f'c:/Users/admin/Desktop/101_강의_데이터/관광객데이터/files/kto_{yymm}.xlsx',
                                header= 1, # 컬럼명의 위치 지정
                                skipfooter=4,
                                usecols="A:G")

            kto['기준년월'] = f'{yy}-{str(mm).zfill(2)}'
            kto_con = kto[~(kto['국적'].isin(continents_list))].reset_index(drop=True)
            kto_con['대륙'] = continents

            kto_con['관광객비율(%)'] = round(kto_con['관광']/kto_con['계']*100, 1)
            kto_con['전체비율(%)'] = round(kto_con['관광']/sum(kto_con['관광'])*100, 1)
            kto_con = kto_con.sort_values('전체비율(%)', ascending=False)
        
            kto_total = pd.concat([kto_total, kto_con], ignore_index=True)
        
        except:
            pass
        
len(kto_total)
```
